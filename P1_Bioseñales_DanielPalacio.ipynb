```python
#Creación de funciones
n = np.arange(0,101,1) #Definicion del intervalo
y = np.sin(np.pi*0.12*n)
y2 = np.cos(np.pi*2*0.03*n)
```

```python
#Operaciones con señales
suma = y + y2
mult = y * y2
```

```python
#Graficación de señales originales superpuestas
plt.figure()
plt.plot(n,y,label='Y1', color='red') #Diferentes plots para que funcione legend
plt.plot(n,y2,label='Y2',color='green',marker='o') #Diferentes plots para que funcione legend
plt.title('Señales originales')
plt.legend()
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.show()
```

```python
#Graficación de señales originales superpuestas
plt.figure()
plt.plot(n,suma,'r--',label='Suma')
plt.plot(n,mult,'g:',label='Multiplicación',)
plt.title('Señales con operaciones')
plt.legend()
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.show()
```

CÁLCULO DE NOTA MÍNIMA, MÁXIMA, MEDIA Y DESVIACIÓN TÍPICA

```python
def f_notas(dic):
  notas = np.array(list(dic.values()),dtype=int)
  datos = {
    "Mínimo": np.min(notas),
    "Máximo": np.max(notas),
    "Media": np.mean(notas),
    "Desviación Std": np.std(notas)}
  serie = pd.Series(datos)
  return datos
```

```python
notas_dict = {
    "Daniel": "5",
    "Ana": "3",
    "Carlos": "4",
    "Laura": "3",
    "Pedro": "4",
    "Sofía": "4",
    "Miguel": "3",
    "Camila": "2",
    "Javier": "3",
    "Lucía": "1"
}

series = f_notas(notas_dict)
print(series)
```

CÁLCULO ÍNDICE DE MASA CORPORAL (IMC)

```python
#Carga de datos con parámetro de separación
bd = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BIOSEÑALES/P1/datos.csv',sep=';')
print(bd.head()) #Mostrar primeros 5 datos
print(bd.tail()) #Mostrar últimos 5 datos
```

```python
#Eliminación primera columna
bd = bd.drop('Unnamed: 0',axis=1, inplace=False)
```

```python
#Calcúlo IMC
bd['IMC'] = bd['Height']/((bd['Weight']/100)**2)
print(bd.head())
```

```python
#Categorización
bd['Categoria'] = bd['IMC'].apply(lambda x: 'Bajo peso' if x < 18.5 else
                                  'Normal' if x < 25 else
                                  'Sobrepeso' if x < 30 else
                                  'Obesidad')
print(bd.head())
```
